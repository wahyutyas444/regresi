# Studi Kasus 1 Data tinggi dan berat badan

```python
# Import library yang diperlukan
from scipy import stats
```

```python
# Data tinggi badan (cm)
x = [150, 160, 170, 180, 190]

# Data berat badan (kg)
y = [50, 55, 65, 70, 80]
```

```python
# Melakukan perhitungan linear regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

# Menampilkan hasil regresi
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Correlation Coefficient (r): {r}")
```

```python
# Buat fungsi linear regression
def myfunc(x):
    return slope * x + intercept
```

# Studi Kasus 2 Data Penjualan

```python
# Import library yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Memanggil dataset dari file Excel
# Pastikan file "Data Penjualan.xlsx" sudah diunggah ke Google Colab
dataset = pd.read_excel("Data_Penjualan.xlsx")

# Pisahkan fitur (x) dan target (y)
x = dataset.iloc[:, :-1].values  # Semua kolom kecuali kolom terakhir sebagai fitur
y = dataset.iloc[:, -1].values   # Kolom terakhir sebagai target
```

```python
# Menampilkan kunci kolom dataset
print("Kolom dataset:", dataset.keys())

# Menampilkan bentuk dataset
print("Bentuk dataset:", dataset.shape)

# Menampilkan isi sebagian dataset
print("Contoh data:")
display(dataset.head())
```

```python
# Split dataset menjadi Training Set dan Testing Set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Tampilkan informasi tentang jumlah data
print(f"Jumlah data training: {len(x_train)}")
print(f"Jumlah data testing: {len(x_test)}")
```

```python
# Fitting Simple Linear Regression pada Training Set
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Menampilkan koefisien regresi
print("Koefisien regresi:", regressor.coef_)
print("Intercept regresi:", regressor.intercept_)
```

```python
# Memprediksi hasil Test-Set
y_pred = regressor.predict(x_test)

# Menampilkan hasil prediksi dan data aktual
print("Hasil prediksi:")
print(y_pred)
print("Data aktual:")
print(y_test)
```

```python
# Visualisasi hubungan antara Biaya Produksi dan Nilai Penjualan
plt.scatter(dataset['BiayaProduksi'], dataset['NilaiPenjualan'], color='blue')
plt.xlabel("Biaya Produksi")
plt.ylabel("Nilai Penjualan")
plt.title("Grafik Biaya Produksi vs Nilai Penjualan")
plt.show()
```

```python
# Visualisasi hasil prediksi pada Training Set
plt.figure(figsize=(10, 8))

# Data observasi
plt.scatter(x_train, y_train, color='blue')

# Hasil prediksi
plt.plot(x_train, regressor.predict(x_train), color='red')

# Judul dan label
plt.title("Biaya Produksi terhadap Penjualan (Training Set)")
plt.xlabel("Biaya Produksi")
plt.ylabel("Nilai Penjualan")

# Menampilkan plot
plt.show()
```

```python
# Visualisasi hasil prediksi pada Test Set
plt.figure(figsize=(10, 8))

# Data observasi
plt.scatter(x_test, y_test, color='blue')

# Hasil prediksi
plt.plot(x_test, regressor.predict(x_test), color='red')

# Judul dan label
plt.title("Biaya Produksi terhadap Penjualan (Test Set)")
plt.xlabel("Biaya Produksi")
plt.ylabel("Nilai Penjualan")

# Menampilkan plot
plt.show()
```
